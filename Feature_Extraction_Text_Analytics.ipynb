# Count and Filter Words

**Practice 1: Count the number of words in text**

```python
num_words = len(text)
print("Number of words:", num_words)
```

```python
# Filter text
stop_words = stopwords.words('english')
print(stop_words) # Notice the stop words are all in lower case
word_tokens = word_tokenize(text)
print(word_tokens)
```

**Practice 2: Fill in the code below to convert the words in word_tokens to lower case and then check whether the words are present in stop_words or not**

```python
filtered_text = []
for word_token in word_tokens:
  # Add code here
  word_token = word_token.lower()
  if word_token not in stop_words:
      filtered_text.append(word_token)
print("Words not in stop_words:", filtered_text)
```

**Practice 3: Modify the code above to only keep words with alphabet letters**

```python
filtered_text = []
for word_token in word_tokens:
    word_token = word_token.lower()
    if word_token not in stop_words and word_token.isalpha():
        filtered_text.append(word_token)
print(filtered_text)
```

# Analyze Sentiment

```python
# Calculate sentiment score with VADER
sid_obj = SentimentIntensityAnalyzer()
# The polarity_scores method of SentimentIntensityAnalyzer object gives a sentiment dictionary
# which contains pos, neg, neu, and compound scores
sentiment_dict = sid_obj.polarity_scores(text) # Notice we feed in the whole text to get a sentiment score
print(sentiment_dict)
print("neg score: ", sentiment_dict['neg'])
print("neu score: ", sentiment_dict['neu'])
print("pos score: ", sentiment_dict['pos'])
print("compound score: ", sentiment_dict['compound'])
```

```python
# Instead of feeding in the whole text (a paragraph), we want to break the paragraph into sentences
# We get a sentiment score (neg, neu, pos, compound) for each sentence
# and then calculate the average score across sentences

# We separate a paragraph into sentences first
sentences = tokenize.sent_tokenize(text)
print(sentences)
```

**Practice 4: Modify the code below to calculate the average neg, neu, pos, compound sentiment scores across sentences**

```python
# Add code here
sid_obj = SentimentIntensityAnalyzer()
text = "Your text goes here."
sentences = tokenize.sent_tokenize(text)
total_neg = 0
total_neu = 0
total_pos = 0
total_compound = 0
for sentence in sentences:
    sid_obj = SentimentIntensityAnalyzer()
    ss = sid_obj.polarity_scores(sentence)
    # Add code here
    total_neg += ss['neg']
    total_neu += ss['neu']
    total_pos += ss['pos']
    total_compound += ss['compound']
num_sentences = len(sentences)
average_neg = total_neg / num_sentences
average_neu = total_neu / num_sentences
average_pos = total_pos / num_sentences
average_compound = total_compound / num_sentences
print("Average neg score: ", average_neg)
print("Average neu score: ", average_neu)
print("Average pos score: ", average_pos)
print("Average compound score: ", average_compound)
```
